## Nettoyage des données 

```python
 # Identification des colonnes avec 0 valeur non-nulle
cols_to_drop = [col for col in data.columns if data[col].count() == 0]
data.drop(columns=cols_to_drop, inplace=True)
print(f"Colonnes supprimées (car entièrement vides) : {cols_to_drop}")

#Il y a aussi des colonnes redondantes ou inutiles pour l'analyse
redundant_cols = ['Ads', 'Ad name', 'Delivery level', 'Attribution setting', 'Objective.1', 'Reporting starts', 'Reporting ends']

# On vérifie si les colonnes existent avant de les supprimer
existing_redundant_cols = [col for col in redundant_cols if col in data.columns]
data.drop(columns=existing_redundant_cols, inplace=True)
print(f"Colonnes redondantes supprimées : {existing_redundant_cols}")
```

```python
# ---TRAITEMENT DES VALEURS MANQUANTES (NaN) ---
# Remplacement des NaN dans les colonnes numériques par 0
numeric_cols = data.select_dtypes(include=np.number).columns
data[numeric_cols] = data[numeric_cols].fillna(0)
print("\nLes valeurs manquantes dans les colonnes numériques ont été remplacées par 0.")
```

```python
# ---CONVERSION DES TYPES DE DONNÉES ---

# Conversion de la colonne 'Day' en format datetime
data['Day'] = pd.to_datetime(data['Day'])
print("\nLa colonne 'Day' a été convertie au format datetime.")

# Conversion des colonnes de comptage en entiers (int)
count_cols = ['Results', 'Reach', 'Impressions', 'Link clicks', 'Clicks (all)', 
              'Landing page views', 'Content views', 'Website content views']

for col in count_cols:
    if col in data.columns:
        data[col] = data[col].astype(int)

print("Les colonnes de comptage ont été converties en type 'integer'.")
```

```python
# --- VÉRIFICATION FINALE ---
print("\n--- Nettoyage Terminé ---")
print("\nNouvelles dimensions du jeu de données :", data.shape)
print("\nInformations après nettoyage :")
data.info()
    
print("\nAperçu des données nettoyées :")
print(data.head())
```

## Effectuons une analyse descriptive des données ( EDA: Exploration Data Analytics)

```python
# --- ANALYSE EXPLORATOIRE (EDA) ---
    
# Calcul des totaux pour les métriques clés
total_spent = data['Amount spent (USD)'].sum()
total_impressions = data['Impressions'].sum()
total_reach = data['Reach'].sum()
total_link_clicks = data['Link clicks'].sum()
    
# Calcul des moyennes agrégées
# On évite la division par zéro si total_impressions ou total_link_clicks est nul
cpm_avg = (total_spent / total_impressions * 1000) if total_impressions > 0 else 0
cpc_avg = (total_spent / total_link_clicks) if total_link_clicks > 0 else 0
ctr_avg = (total_link_clicks / total_impressions * 100) if total_impressions > 0 else 0
    
print("\n--- Performance Globale de la Campagne ---")
print(f"Montant total dépensé : ${total_spent:,.2f}")
print(f"Impressions totales : {total_impressions:,}")
print(f"Clics sur le lien totaux : {total_link_clicks:,}")
print("---------------------------------------------")
print(f"CPM moyen (Coût pour 1000 impressions) : ${cpm_avg:,.2f}")
print(f"CPC moyen (Coût par clic sur le lien) : ${cpc_avg:,.2f}")
print(f"CTR moyen (Taux de clics sur le lien) : {ctr_avg:,.2f}%")
```

Performance Globale :
268 dollars dépensés pour générer 949 clics. C'est le résultat principal.
CPC moyen de 0.28$ : C'est notre coût d'acquisition de trafic. 
CTR moyen de 4.03% : C'est un excellent taux de clics. Cela signifie que les publicités sont visuellement attrayantes ou que le ciblage est très pertinent pour l'audience qui les voit.

### Analyse des Performances par Campagne
L'ensemble de données contient plusieurs campagnes (Campaign name).

```python
# --- ANALYSE PAR CAMPAGNE ---
print("\n--- Analyse des Performances par Nom de Campagne ---")

# Utilisons groupby() pour agréger les données par campagne
campaign_performance = data.groupby('Campaign name').agg({
    'Amount spent (USD)': 'sum',
    'Impressions': 'sum',
    'Link clicks': 'sum'
}).sort_values(by='Amount spent (USD)', ascending=False)

# Calcul des métriques CPC et CTR pour chaque campagne
campaign_performance['CPC'] = campaign_performance['Amount spent (USD)'] / campaign_performance['Link clicks']
campaign_performance['CTR'] = (campaign_performance['Link clicks'] / campaign_performance['Impressions']) * 100

# Remplacons les NaN ou infinis si une campagne n'a pas eu de clics ou d'impressions
campaign_performance.replace([np.inf, -np.inf], np.nan, inplace=True)
campaign_performance.fillna(0, inplace=True)


print(campaign_performance)
```

## Analyse des Performances par Segments Démographiques (Âge et Genre)
C'est une analyse cruciale pour optimiser le ciblage. Quelle tranche d'âge ou quel genre réagit le mieux à nos publicités ?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration des graphiques
sns.set_style('whitegrid')

# --- ANALYSE PAR SEGMENTS DÉMOGRAPHIQUES ---

# Performance par Genre
gender_performance = data.groupby('Gender')['Amount spent (USD)'].sum().sort_values(ascending=False)
print("\n--- Dépenses par Genre ---")
print(gender_performance)

# Visualisation des dépenses par genre
plt.figure(figsize=(10, 6))
sns.barplot(x=gender_performance.index, y=gender_performance.values)
plt.title('Dépenses Publicitaires Totales par Genre')
plt.ylabel('Montant dépensé (USD)')
plt.xlabel('Genre')
plt.show()


# Performance par Tranche d'Âge
age_performance = data.groupby('Age').agg({
    'Amount spent (USD)': 'sum',
    'Link clicks': 'sum'
}).sort_values(by='Amount spent (USD)', ascending=False)

# Calcul du CPC par tranche d'âge
age_performance['CPC'] = age_performance['Amount spent (USD)'] / age_performance['Link clicks']
age_performance.fillna(0, inplace=True)

print("\n--- Performance par Tranche d'Âge ---")
print(age_performance)

# Visualisation du CPC par tranche d'âge
plt.figure(figsize=(12, 7))
sns.barplot(x=age_performance.index, y=age_performance['CPC'], order=age_performance.sort_values('CPC').index)
plt.title('Coût par Clic (CPC) Moyen par Tranche d\'Âge')
plt.ylabel('CPC Moyen (USD)')
plt.xlabel('Tranche d\'Âge')
plt.xticks(rotation=45)
plt.show()
```

#### tranche d'âge a le CPC le plus bas, C'est souvent l'audience la plus rentable pour l'acquisition de trafic.

```python
# --- ANALYSE TEMPORELLE ---

# Agréger les données par jour
daily_performance = data.groupby('Day').agg({
    'Amount spent (USD)': 'sum',
    'Link clicks': 'sum'
}).reset_index()

print("\n--- Performance Journalière ---")
print(daily_performance.head())

# Visualisation des dépenses journalières
plt.figure(figsize=(15, 7))
plt.plot(daily_performance['Day'], daily_performance['Amount spent (USD)'], marker='o', linestyle='-')
plt.title('Évolution des Dépenses Publicitaires Journalières')
plt.ylabel('Montant dépensé (USD)')
plt.xlabel('Date')
plt.grid(True)
plt.show()

# Visualisation des clics journaliers
plt.figure(figsize=(15, 7))
plt.plot(daily_performance['Day'], daily_performance['Link clicks'], marker='o', linestyle='-', color='green')
plt.title('Évolution du Nombre de Clics Journaliers')
plt.ylabel('Nombre de Clics sur le lien')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```

#### Analyse par Ensemble de Publicités (Ad Set Name)
Quel est l'ensemble de publicités (Ad Set Name) qui a le CPC le plus bas et/ou le CTR le plus élevé ? 
C'est le gagnant de notre A/B test.

```python
# --- ANALYSE PAR ENSEMBLE DE PUBLICITÉS (AD SET) ---
print("\n--- Analyse des Performances par Ensemble de Publicités ---")

ad_set_performance = data.groupby('Ad Set Name').agg({
    'Amount spent (USD)': 'sum',
    'Impressions': 'sum',
    'Link clicks': 'sum'
}).sort_values(by='Amount spent (USD)', ascending=False)

# Calcul des métriques CPC et CTR pour chaque Ad Set
ad_set_performance['CPC'] = ad_set_performance['Amount spent (USD)'] / ad_set_performance['Link clicks']
ad_set_performance['CTR'] = (ad_set_performance['Link clicks'] / ad_set_performance['Impressions']) * 100

ad_set_performance.fillna(0, inplace=True)

print(ad_set_performance)

# Visualisation
plt.figure(figsize=(12, 8))
ad_set_performance['CPC'].sort_values().plot(kind='barh', color='skyblue')
plt.title('Coût par Clic (CPC) par Ensemble de Publicités')
plt.xlabel('CPC Moyen (USD)')
plt.ylabel('Ensemble de Publicités')
plt.grid(True, axis='x')
plt.show()
```

### Croisement des Données Démographiques
Nous savons que les femmes de 55+ sont la cible. Mais est-ce les femmes de 55-64 ans ou les femmes de 65+ qui sont les plus performantes ?
En comparant CPC_female et CPC_male pour chaque tranche d'âge, pouvons-nous confirmer que le ciblage féminin est plus efficace (CPC plus bas) ?

```python
# --- ANALYSE DÉMOGRAPHIQUE CROISÉE (PIVOT TABLE) ---

# Création d'un tableau croisé dynamique pour analyser le CPC par âge et genre
demographic_pivot = pd.pivot_table(
    data,
    values=['Amount spent (USD)', 'Link clicks'],
    index='Age',
    columns='Gender',
    aggfunc='sum',
    fill_value=0
)

# Calcul du CPC dans le tableau croisé
demographic_pivot['CPC_female'] = demographic_pivot[('Amount spent (USD)', 'female')] / demographic_pivot[('Link clicks', 'female')]
demographic_pivot['CPC_male'] = demographic_pivot[('Amount spent (USD)', 'male')] / demographic_pivot[('Link clicks', 'male')]

# Remplacer les infinis par 0 pour une meilleure lisibilité
demographic_pivot.replace([np.inf, -np.inf], 0, inplace=True)


print("\n--- Analyse Croisée du CPC par Âge et Genre ---")
# On affiche uniquement les colonnes CPC pour la clarté
print(demographic_pivot[['CPC_female', 'CPC_male']])
```

Les Gagnants et les Perdants de l'A/B Test


Le Grand Gagnant : HumanCustumer Group A
Regardez le tableau : cet ensemble de publicités a un CPC de 0.18 dollars. C'est de loin le plus bas et le plus performant de tous !
Problème : Il n'a reçu que 1.67 dollars de budget. C'est une opportunité manquée massive. Le meilleur groupe de test a été à peine utilisé.


Le Bon Élève : HumanCustumer Group B
Avec un CPC de 0.25 dollars, c'est le deuxième meilleur groupe et celui qui a reçu le plus de budget (avec "Brand + Product Group A").
Son CTR de 6.35% est excellent, le plus élevé parmi les groupes qui ont eu un budget significatif.
Conclusion : Cette approche "HumanCustumer" (qui met probablement en avant des personnes/témoignages) est clairement plus efficace que l'approche "Brand + Product".


Le Moyen : Brand + Product Group A
Un CPC de 0.28 dollars et un CTR de 3.08%. C'est la performance "moyenne" de la campagne. Il fonctionne, mais il est moins efficace que les groupes "HumanCustumer".


Le Grand Perdant : Brand + Product Group B
Avec un CPC de 0.54 dollars, cet ensemble est deux fois plus cher que la moyenne de la campagne. Chaque clic coûte deux fois plus cher !
Le budget de 27$ dépensé ici aurait pu être alloué à HumanCustumer Group A pour générer trois fois plus de clics.

### Interprétation de l'Analyse Croisée : CPC par Âge et Genre

Le tableau CPC_female vs CPC_male est très instructif.
Pour les femmes (la cible principale) : Le CPC est assez stable et bon pour les 35+ (entre 0.24 et 0.29).
Pour les hommes : Les résultats sont beaucoup plus erratiques. Le CPC est excellent pour les 25-34 ans
